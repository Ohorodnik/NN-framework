```python
lb =  LabelBinarizer()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=1
)
Y_train_ = Y_train
Y_train = lb.fit_transform(Y_train)
```
